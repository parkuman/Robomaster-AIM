## The Basics:
### Variables:
‘Variables’ are given values that are then stored in the computers memory and able to be called and manipulated. For example:


```python
a = 2
```

Since we are assigning ‘a’ the value of 2, the equals sign is called the ‘assignment operator’. This is a rather simple example for assigning values, a more complex one could be as follows:

```python
x_var = [2,3,5,7,9]
```

In this example a list of numbers were assigned the name ‘x_var’. Naming variables specifically will help the reader understand the code. 

### Print Functions:
Now that we have stored values in the computers memory, we can call these values into various functions, one of which being the print function. The print function allows the arguments given to be displayed on the screen neatly. For example:


```python
print('Hello World')
print(a)
```

As evident in the example above, python reads in order, meaning a variable must be defined before it is called. In print functions, strings of text and values can be interchanged. It is important to remember that for a print statement to print out a string of text it must be in quotations. An example of using text and values in a print statement could be as follows:

```python
print('The answer to the eqaution 1+1 is',a)
```

## Formatting
As seen in the Print Functions section, answers can be formatted and printed in the order in which you wish them to appear. However it is important to understand how to properly, round and format mathematically answers such that they are legible and useable. There are several ways to round in python. The method shown is the easiest to understand. It is shown as follows:

```python
num = 1/3
rounded_num = round(num,3)
print(rounded_num)
```

This method can be made shorter without using an intermediate variable and instead using the rounded function within the print function. In addition to proper formatting an important tool in making your code easy to follow is comments. To comment out in python means to place text after a # sign. By doing this, python will not 'read' what is written. For example:

```python
x= 1 #defining the variable
y = 2*x #y is twice x

#now we print the sum of x and y
print(x+y)
```

## Loops
An important part of coding is being able to do tasks multiple times without user input. Looping a function can result in this. A very basic example of a loop is given below. When preforming loops it is important to remeber to indent and use a colon to indicate the loop or various loops.

```python
radii = [0.7,1.1,1.4,0.88]
n = len(radii)
for index in range(n):
    print("Element", index, "has value",radii[index])
```

A more complicated example of a loop is shown below. This example uses all of the information provided up to this point as well as some built in python functions. The code is very well documented and is easy to follow. 

```python
#Question: Find the largest integer divisor of 124327

n = 124327 #define n as dividend

nrange = list(range(1,n)) #create list of possible divisors

divisorList = [] #create divisor list which we will add divisors to

for number in nrange:
    if n % number == 0: #determine which numbers from list result in 0 remainder
        divisorList.append(number) #append divisor list with numbers of 0 remainder
print(max(divisorList)) #print max divisor
```

## Virtual Environments
Collaborating on a web based python project with github is useful for completing large projects. It can also be useful to isolate the project in a virtual enviornment, this allows users to download packages and work on the project without effecting the main project. In essence, your vitual enviornment is a copy of python in which you can import whichever packages you choose and work on a shared project without mixing up libraries with your own projects.

The setup of such an enviornment in straight forward. 

### Windows

1) Install Python
 - By means of following the prompts on the screen, download and install whichever version of python to your computer

2) Add Python to Path 

- This can be done several ways. The first, is when following the python installation prompts, there is an option to add said version of python to PATH, simply select that. Otherwise, take note of where python was downloaded to on your computer (should be user/AppData/Local/Programs/Python/python36) copy this address, and manually add to Path. Also to be added to path is user/AppData/Local/Programs/Python/python36/Scripts.

- Some notes, the AppData folder may be hidden, select view, hidden icons. To find PATH on your computer simply type into the search bar, edit enviornment variables for your account, select this option, select path, new, then paste the copied address.

3) Get pip
- In the console, cmd, paste the following command: python get-pip.py

4) Creating your own virtual enviornment
- Check python is installed, in your terminal paste; pip -h
- Install the virtual enviornment package, in your terminal paste; pip install virtualenv
- Next create the virtual enviornment naming it whatever you want, by pasting into your terminal; virtualenv name
- To activate your virtual enviornment, paste into the terminal; name/Scripts/activate
- When you are finished working in your virtual enviornment simply write; deactivate in your console.
